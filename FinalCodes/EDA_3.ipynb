```python
# Create a column to hold number of orders in the last 14 days
from collections import defaultdict
from datetime import timedelta

order_counts_14d = []

# Group by customer
for name, group in df.groupby('customer_name'):
    group = group.sort_values('order_date')
    order_dates = group['order_date'].tolist()
    counts = []

    for i in range(len(order_dates)):
        current_date = order_dates[i]
        window_start = current_date - timedelta(days=14)
        count = sum((window_start <= d <= current_date) for d in order_dates[:i+1])
        counts.append(count)

    order_counts_14d.extend(counts)

df['orders_last_14_days'] = order_counts_14d

# For each customer, calculate the gap in days between orders
df['days_since_last_order'] = df.groupby('customer_name')['order_date'].diff().dt.days


# Remove NaNs (first order won't have a previous one)
#df_filtered = df.dropna(subset=['days_since_last_order'])

# Average reorder interval
avg_reorder_time = df.groupby('customer_name')['days_since_last_order'].mean().reset_index()


df
```

Use negative profit or low invoiced_quantity vs quantity as proxy for understocking.

If quantity ordered > invoiced quantity → Stockout occurred.

These SKUs need extra safety stock

We’ll look at two loss signals:

quantity > invoiced_quantity → Order was not fully fulfilled → Likely a stockout

profit < 0 → You sold at a loss → Possibly discounting due to overstock or missed margin due to rush purchases

```python
df['stockout_flag'] = (df['quantity'] > df['invoiced_quantity']).astype(int)
df
```

```python
df['unfulfilled_qty'] = df['quantity'] - df['invoiced_quantity']
df['unfulfilled_qty'] = df['unfulfilled_qty'].apply(lambda x: x if x > 0 else 0)
df
```

```python
stockout_summary = (
    df
    .groupby('item_name')
    .agg(
        stockout_events=('stockout_flag', 'sum'),
        total_orders=('order_id', 'count'),
        total_unfulfilled=('unfulfilled_qty', 'sum'),
        avg_unfulfilled_per_event=('unfulfilled_qty', lambda x: x[x > 0].mean()),
        lost_profit=('profit_y', lambda x: x[x < 0].sum())
    )
    .sort_values(by='stockout_events', ascending=False)
)
stockout_summary
```

```python
stockout_summary['stockout_rate'] = stockout_summary['stockout_events'] / stockout_summary['total_orders']
stockout_summary['safety_multiplier'] = 1 + stockout_summary['stockout_rate'] * 1.5  # tunable factor
stockout_summary
```


1 : This is your baseline — you always need to plan for what you forecast.

stockout_rate : Proportion of orders where stock was unavailable. A direct risk indicator.

× 1.5 : A severity multiplier — how aggressively you want to compensate.

Imagine you forecast 100 SKUs, and you’re 100% accurate for 95 low-volume ones, but miss badly on 5 high-revenue ones.

➡️ Total forecast accuracy = 95% ✅
➡️ Business performance = ❌ disaster (you missed the big money-makers)
So:

✅ You're forecasting better where it matters

✅ You reduce risk of lost revenue on best-sellers

✅ You align model performance with business outcomes

🔁 1. Revenue-Weighted Loss Function (Model Training)
Train your forecasting model using a custom loss like:

Weighted_MAE = sum( abs(actual - forecast) * sku_weight ) / sum(sku_weights)
So mistakes on high-revenue SKUs are penalized more

📋 2. Weighted Evaluation Metrics
Instead of reporting raw MAE or RMSE per SKU:


weighted_error = (errors * sku_weights).sum()
This is a better business-aligned KPI for model performance.


       Benefit	                                 Why It Works
Business-aligned accuracy	     Reduces errors where revenue impact is highest
Smarter model optimization	     Prioritizes the SKUs that matter
Scalable resource allocation	 More effort spent on high-impact predictions
Better customer experience	     Ensures stock availability on top-demand items
Lean inventory	                 Low-weight SKUs don’t get excess safety stock

📦 Why Warehouse-Level Forecasting Matters

    Challenge	                            What Happens Without Per-Warehouse Forecasting	                           Why It’s Bad
Regional demand patterns	      Demand varies by location	                                       You overstock in some regions, understock in others
Shipping & fulfillment latency	  Customers get late/missed deliveries	                           Poor satisfaction
Warehouse-specific stockouts	  You may have stock in WH1 but not WH2	                           You miss easy wins
Local events/seasonality	      Warehouse A may have a festival, B does not	                   Blind forecasting fails

⚖️ Why Use invoiced_quantity Instead of quantity (Ordered Qty)?
Because inventory planning should be based on what was actually fulfilled, not just what was requested.

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from lightgbm import early_stopping, log_evaluation

# Categorical columns
cat_cols = ['item_name', 'warehouse_name', 'dayofweek', 'month', 'is_weekend']
feature_cols = [
    'item_name',                   
    'warehouse_name',            
    'dayofweek',               
    'month',                     
    'day',
    'is_weekend',             
    'week_of_month',
    'days_since_month_start',
    'avg_last_7d',             
    'avg_last_14d',            
    'std_last_7d',             
    'days_since_last_sale',  
    'safety_multiplier'       
]
X = features[feature_cols]
y = features['invoiced_quantity']

# Time-based split (no shuffle)
X_train, X_val, y_train, y_val = train_test_split(X, y, shuffle=False, test_size=0.2)

# LightGBM datasets
train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=cat_cols)
val_data = lgb.Dataset(X_val, label=y_val, categorical_feature=cat_cols)

# Train the model with early stopping using callbacks
model = lgb.train(
    params={
        'objective': 'regression',
        'metric': 'rmse',
        'learning_rate': 0.05,
        'num_leaves': 31,
        'verbose': -1
    },
    train_set=train_data,
    valid_sets=[train_data, val_data],
    num_boost_round=500,
    callbacks=[
        early_stopping(stopping_rounds=50),
        log_evaluation(period=50)
    ]
)
```
